```python
import pytesseract
from PIL import Image
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

print("請上傳圖片：")
uploaded = files.upload()

if uploaded:
    fn = list(uploaded.keys())[0]
    file_bytes = np.frombuffer(uploaded[fn], np.uint8)
    img = cv2.imdecode(file_bytes, cv2.IMREAD_GRAYSCALE)

    _, thresh = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY_INV)
    kernel = np.ones((1,1), np.uint8)
    dilated_img = cv2.dilate(thresh, kernel, iterations=1)


    final_processed = cv2.bitwise_not(dilated_img)


    custom_config = r'--oem 3 --psm 6'
    text = pytesseract.image_to_string(final_processed, lang='chi_tra+eng', config=custom_config)

    print("OCR 辨識結果:")
    print("-" * 40)
    print(text)
```
