```python
import pytesseract
from PIL import Image
import cv2
from google.colab import files

print("請上傳你要辨識的圖片：")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

image = Image.open(file_name)

text = pytesseract.image_to_string(image, lang='eng', config='--psm 10')

print("OCR 辨識結果:")
print("-" * 40)
print(text)
```
